# 최소 자승법



```python
x = np.array([1, 2, 3])   # 2x+ 1
y = np.array([3, 5, 8])

x_ = np.sum(x)
xy_ = np.sum(x*y)
y_ = np.sum(y)
x2_ = np.sum(x**2)
n = x.size  # len(x)
x_, xy_, y_, x2_, n

A = np.array([[x2_, x_],
              [x_,  n]])
Y = np.array([[xy_],
              [y_]])
print(A)
np.matmul(np.linalg.inv(A), Y)
```

# 뉴턴 알고리즘


```python
def fun(x) :
    return x**2 + 4*x + 2

def d_fun(x) :
    return 2*x + 4

x = 2

for i in range(10) :
    x = x - fun(x) / d_fun(x)
    print(x)  
    
```

```python
#cos(x) - 3x^2  = 0

def fun(x) :
    return np.cos(x) - 3*x**2

def d_fun(x) :
    return -np.sin(x) - 6*x

x = 10

for i in range(20) :
    x = x - fun(x) / d_fun(x)
    print(x) 
```
