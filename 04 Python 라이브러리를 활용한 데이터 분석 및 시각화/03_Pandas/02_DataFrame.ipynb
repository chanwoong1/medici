# DataFrame
- 2차원 행렬 데이터를 행렬 형태가 아닌 테이블 형태로 관리할 수 있도록 제공되는 요소
- DataFrame의 컬럼 하나는 하나의 Series로 구성되어 있으며 다수의 Series들을 모아 하나의 DataFrame으로 관리한다.

```python
import numpy as np
import pandas as pd
```

## 데이터 프레임 생성
- 2차원 리스트
- 리스트를 가지고 있는 딕셔너리
- 외부 파일(csv, 엑셀, spss 등)
- 웹 문서의 table tag
- 데이터 베이스의 테이블
- 기타 등등

```python
# 2차원 리스트를 통한 생성
# 내부의 각 리스트의 인덱스가 일치하는 것 끼리 모아 Series를 만든다.
# 데이터 개수가 다르게 되어 있을 경우에는 비어있는 곳은 결측치로 채워진다.
list1 = [
    [1, '남자', 90, 80, 70],
    [2, '여자', 91, 81, 71],
    [3, '남자', 92, 82, 72],
    [4, '여자', 93, 83, 73],
    [5, '남자', 94, 84],
    [6, '여자', 95, 85, 75, 65]
]

df1 = pd.DataFrame(list1)
df1

```

```python
# 컬럼 이름에 사용한 이름
column_list = ['번호','성별','국어','영어','수학','과학']
# 인덱스에 사용할 이름
index_list = ['학생1','학생2','학생3','학생4','학생5','학생6']

df2 = pd.DataFrame(list1, columns= column_list, index= index_list)
df2
```

```python
# 리스트를 가지고 있는 딕셔너리를 통한 생성
# 딕셔너리에 지정한 이름이 컬럼의 이름으로 결정된다.
# 데이터의 수가 다르게 되어 있으면 오류가 발생한다.
data1 = {
    '번호' : [1, 2, 3, 4],
    '성별' : ['남자', '여자', '남자', '여자'],
    '국어' : [90, 91, 92, 93],
    '영어' : [80, 81, 82, 83],
    '수학' : [70, 71, 72, 73],
#     '과학' : [60, 61, 62, 63],
#     '한국사' : [50, 51, 52]
}

df3 = pd.DataFrame(data1)
df3
```

```python
# 딕셔너리를 가지고 있는 리스트
# 각 딕셔너리에 이름이 똑같은 것들까지 모아 하나의 Series로 구성한다.
list3 = [
    {'번호' : 1, '이름' : '홍길동', '국어' : 100},
    {'번호' : 2, '이름' : '김길동', '국어' : 101},
    {'번호' : 3, '이름' : '최길동', '국어' : 102},
    {'번호' : 4, '이름' : '박길동', '국어' : 103},
    {'이름' : '황길동', '국어' : 104, '번호' : 5},
    {'번호' : 6, '이름' : '고길동', '귝어' : 100},
]

df4 = pd.DataFrame(list3)
df4
```

```python
# csv 파일
# 데이터를 쉼표(,)로 구분한 데이터 양식
# 데이터 프레임 생성시 Encoding 오류가 발생하거나 생성된 데이터 프레임의 한글 데이터가 깨질 경우
# pandas는 기본 인코딩이 utf-8로 설정되어 있다. 허나 한국에서 엑셀을 통해 작업한 csv 파일은 인코딩이 euc-kr로 되어있다.
# csv 파일을 통해 생성시 encoding 오류가 나거나 한글이 깨져서 나오면 encoding을 euc-kr로 설정해서 시도해본다.

df1 = pd.read_csv('data/grade.csv', encoding='euc-kr')
df1
```

# 데이터의 종류
- 레이블형, 카테고리형 : 값의 종류가 정해져 있는 값(혈액형, 성별, 지하철 호선)
- 범위형 : 값의 범위(최소와 최대가 존재)하는 값(나이, 키, 몸무게 등)
- 데이터의 종류에 따라 통계값을 해석해야 한다.

```python
# 요약 통계 정보
# 데이터 프레임 내의 숫자 타입 걸럼들만 모아 요약 통계 정보를 제공한다.
df1.describe()
```

```python
# 범위형 데이터에 대한 요약 통계 정보
# count: 결측치를 제외한 데이터의 개수
# mean : 결측치를 제외한 평균
# std : 결측치를 제외한 표준편차
# min : 결측치를 제외한 최소값
# 25% : 최대 최소 범위 안에 25% 위치의 값(존재하지 않을 수도 있다.)
# 59% : 최대 최소 범위 안에 50% 위치의 값(존재하지 않을 수도 있다.)
# 75% : 최대 최소 범위 안에 75% 위치의 값(존재하지 않을 수도 있다.)
# max : 결측치를 제외한 최대값

# min과 max를 통해 정상 범위에 벗어나는 데이터가 있는지 확인한다.
df1[['Age','Fare']].describe()
```

```python
# 레이블형 데이터
# count : 결측체를 제외한 데이터의 개수
# unique : 값의 종류
# top : 가장 많이 저장되어 있는 값
# freq : 가장 많이 저장되어 있는 값이 몇 개가 저장되어 있는지
# 요약 통계 정보는 숫자 컬럼인 경우에는 범위형 데이터로 취급해 보여주게 되고 문자열이라고 하더라도 어떠한 데이터가 저장되어 있는지 알 수 없기 때문에 이상치 존재 여부를 확인하기가 어렵다.

df1['Embarked'].describe()
```

```python
# value_counts : Series 내에 어떠한 값이 몇 개씩 저장되어 있는지 결과를 반환해준다

display(df1['Pclass'].value_counts())
display(df1['Embarked'].value_counts())
```

## 특정 위치의 값에 접근한다.

```python
df1 = pd.read_csv('data/grade.csv', encoding = 'euc-kr')
df1.index = ['학생1', '학생2', '학생3', '학생4', '학생5']
df1
```

```python
# 특정 컬럼을 가져온다.
c1 = df1['국어']
c1
```

```python
# 특정 행을 추출한다.(index)
r1 = df1.loc['학생3']
r1
```

```python
# 특정 행을 가져온다(순서, 0부터 시작)
r2 = df1.iloc[2]
r2
```

```python
# 원하는 위치의 값을 가져온다.

v1 = df1['국어']['학생3']
v1
```

```python
v2 = df1['국어','학생3']
v2   # 오류 발생
```

```python
# 행, 열 접근
v3 = df1.loc['학생3']['국어']
v3
```

```python
# 순서 (3번째 행, 4번째 열)
v4 = df1.iloc[2][3]
v4
```

```python
v5 = df1.iloc[2,3]
v5
```

```python
# 열 행 접근 방식으로 값을 설정한다.
df1['국어']['학생3'] = 1000
df1
```

```python
# 행 열 접근 (index)
df1.loc['학생3']['영어'] = 2000
df1     # 바뀌지 않았다
```

```python
# 이전에 본 표기 기법은 그것에 해당하는 것들을 추출하여 새로운 데이터 프레임을 생성해서 반환한다.
# 아래의 표기 기법은 해당 데이터 프레임에 직접 접근을 한다.
df1.loc['학생3','영어'] = 2000
df1
```

```python
df1.iloc[2, 5] = 3000
df1
```
