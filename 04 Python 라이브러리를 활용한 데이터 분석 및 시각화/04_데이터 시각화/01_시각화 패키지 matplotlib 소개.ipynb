- Matplotlib는 파이썬에서 자료를 시각화하는 패키지
- 라인 플롯, 스캐터 플롯, 컨투어 플롯, 서피트 플롯, 바 차트, 히스토그램, 박스 플롯 등 다양한 그래프를 그릴 수 있다.
- https://matplotlib.org/stable/gallery/index.html

```python
import matplotlib.pyplot as plt
import matplotlib as mpl

# 그래프를 그리기 위한 기본 설정
# 폰트 설정 - 윈도우
plt.rcParams['font.family'] = 'Malgun Gothic'

# 그래프의 크기
plt.rcParams['figure.figsize'] = 12, 6

# 글자 크기
plt.rcParams['font.size'] = 14

# 폰트 설정 시 - 기호 깨는거 방지하기
plt.rcParams['axes.unicode_minus'] = False
```

# 라인 플롯
- 가장 간단한 그래프
- 데이터가 시간, 순서 등에 따라 어떻게 변화되는지 보여주기 위해 사용한다.

```python
plt.title('선 그래프' )
# y축에 설정될 값만 설정하면 x축은 값의 개수로 설정된다.
plt.plot([10, 20, 30, 40], [1, 4, 9, 16])
plt.show()
```

# 각 축의 이름 설정

```python
plt.title('선 그래프' )
# 리스트를 두 개 설정하면 x축, y축이 된다.
plt.plot([10, 20, 30, 40], [1, 4, 9, 16])

# x축 이름
plt.xlabel('x 축')
# y축 이름
plt.ylabel('y 축')

plt.show()
```

# 그래프의 스타일 지정

```python
plt.title('선 그래프')
# 리스트를 두 개 설정하면 x축, y축이 된다.
plt.plot([10, 20, 30, 40], [1, 4, 9, 16], 'rs--')

# x축 이름
plt.xlabel('x 축')
# y축 이름
plt.ylabel('y 축')

plt.show()
```

# 그래프의 스타일
- '색상 마커종류 선종류' 형태로 작성한다 (예 , rs--)


## 색상
- https://matplotlib.org/stable/gallery/color/named_colors.html#sphx-glr-gallery-color-named-colors-py


## 마커
- . , o v ^ < > 1 2 3 4 s p * h H + x D d


## 선 스타일

|문자|의미|
|---|---|
| - |실선|
| -- |dash|
| -. |점선과 dash|
| : |점선|

```python
plt.title('선 그래프')
# 리스트를 두 개 설정하면 x축, y축이 된다.
plt.plot([10, 20, 30, 40], [1, 4, 9, 16], 'm^-.')

# x축 이름
plt.xlabel('x 축')
# y축 이름
plt.ylabel('y 축')

plt.show()
```

# 상세 스타일
- c : 선 색상
- lw : 선 굵기
- ls : 선 스타일
- marker : 마커 종류
- ms : 마커 크기
- mec : 마커 색상
- mew : 마커 선 굵기
- mfc : 마커 내부 색상

```python
plt.title('선 그래프')
# 리스트를 두 개 설정하면 x축, y축이 된다.
plt.plot([10, 20, 30, 40], [1, 4, 9, 16], c = 'b', lw = 5, ls = ':', marker = 'o', ms = 15, mec = 'g', mew = 5, mfc = 'r')

# x축 이름
plt.xlabel('x 축')
# y축 이름
plt.ylabel('y 축')

plt.show()
```

# 범위 지정

```python
plt.title('선 그래프')
# 리스트를 두 개 설정하면 x축, y축이 된다.
plt.plot([10, 20, 30, 40], [0.81, 0.91, 0.83, 0.92], 'rs--')

# 각 축의 범위를 설정한다.
plt.ylim(0, 1)
plt.xlim(0, 50)

# 각 축의 눈금을 설정한다.
plt.yticks([0, 0.5, 1])
plt.xticks([0, 20, 40, 60])

# x축 이름
plt.xlabel('x 축')
# y축 이름
plt.ylabel('y 축')

plt.show()
```

# 그리드 설정
- 과거 버전에서는 그리드가 나오는 것이 기본이었다.
- 현재는 그리드가 나오지 않는 것이 기본이다.

```python
plt.title('선 그래프')
# 리스트를 두 개 설정하면 x축, y축이 된다.
plt.plot([10, 20, 30, 40], [1, 4, 9, 16], c = 'b', lw = 5, ls = ':', marker = 'o', ms = 15, mec = 'g', mew = 5, mfc = 'r')

# x축 이름
plt.xlabel('x 축')
# y축 이름
plt.ylabel('y 축')

plt.grid()
# 만약 grid가 기본으로 나오는 나위 버전일 경우
# plt.grid(False)

plt.show()
```

# 범례

```python
import numpy as np
```

```python
# 시작값, 종료값, 숫자개수 : 시작값에서부터 종료값까지 총 숫자개수 만큼 숫자를 가지고 있는 ndarray를 생성한다.
X = np.linspace(-np.pi, np.pi, 256)
y1 = np.cos(X)
y2 = np.sin(X)

plt.title('범례 표시하기')
plt.plot(X, y1, ls = '--', label = 'cosine')
plt.plot(X, y2, ls = ':', label = 'sine')

# 각 그래프의 label 값을 활용하여 범례를 표시한다
# loc : 범례의 위치, 생략시 자동
# 0 (best, 기본), 1 (우상단), 2 (좌상단), 3 (좌하단), 4 (우하단), 5 (우측), 6 (좌측), 7 (우중앙), 8 (중하단), 9 (중상단), 10 (정중앙)
plt.legend(loc = 1)

plt.show()
```
