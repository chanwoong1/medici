```python
import numpy as np
import matplotlib.pyplot as plt

# 그래프를 그리기 위한 기본 설정
# 폰트 설정 - 윈도우
plt.rcParams['font.family'] = 'Malgun Gothic'

# 그래프의 크기
plt.rcParams['figure.figsize'] = 12, 6

# 글자 크기
plt.rcParams['font.size'] = 14

# 폰트 설정 시 - 기호 깨는거 방지하기
plt.rcParams['axes.unicode_minus'] = False
```

# 선 그래프
- 어떠한 데이터 흐름에 따라 변화되는 데이터를 표현할 때 사용한다.

```python
X = [1, 2, 3]
y = [2, 3, 1]

plt.title('선 그래프')
plt.plot(X, y)

# 각 축의 값의 종류와 동일한 값을 넣어줘야 한다.
# 선 그래프에 어울리는 데이터는 숫자인 경우이기 때문에 숫자로 설정하면 된다.
plt.xticks([1, 2, 3], ['가', '나', '다'])
plt.yticks([0, 2, 4], ['하', '중', '상'])

# 각 축의 이름
plt.xlabel('x축')
plt.ylabel('y축')

plt.show()
```

# 막대 그래프 
- 데이터들의 양을 비교하기 위해서 사용

```python
X = [1, 2, 3]
y = [2, 3, 1]

plt.title('막대 그래프')
plt.bar(X, y)

# 각 축의 값의 종류와 동일한 값을 넣어줘야 한다.
# 선 그래프에 어울리는 데이터는 숫자인 경우이기 때문에 숫자로 설정하면 된다.
plt.xticks([1, 2, 3], ['가', '나', '다'])
plt.yticks([0, 2, 4], ['하', '중', '상'])

# 각 축의 이름
plt.xlabel('x축')
plt.ylabel('y축')

plt.show()
```

# 스템 플롯
- 막대 그래프와 유사하지만 폭이 없는 그래프이다.
- 이산 확률 함수나 자기 상관 관계를 묘사할 때 사용한다.

```python
x = np.linspace(0.1 , 2 * np.pi, 10)
y = np.cos(x)

plt.title('스템 플롯')
plt.stem(x, y, '-.')

plt.show()
```

# 파이 차트
- 카테고리 별 값의 상대적인 비교를 해야 할 때
- 막대 그래프는 양을 비교하는 것이고 파이차트는 비율을 비교한다

```python
values = [15, 30, 40, 10]
labels = ['값1', '값2', '값3', '값4']
colors = ['yellowgreen', 'gold', 'lightskyblue', 'white']

# 조각의 튀어나옴 정도
explode = [0, 0.1, 0, 0]

plt.pie(values, labels = labels, colors = colors, explode = explode, shadow = True, counterclock=False, autopct = '%.2f%%')
plt.title('파이 차트', color = 'white')
plt.show()
```

# 히스토그램
- 막대그래프와 유사해 보이지만 구간별 양을 비교하는 용도로 사용한다.

```python
x = np.random.randn(1000)

plt.title('히스도그램')
plt.hist(x, bins = 30)
plt.show()
```

# 산점도
- 2차원 데이터의 상관관계를 파악할 수 있다
- 머신러닝에서는 데이터를 산점도로 그려서 분포를 파악하고 이를 통해 분류 가능 여부를 판단할 수 있다. 중펍되는 구간이 너무 많으면 어떤 학습 알고리즘을 사용해도 성능이 매우 떨어질 수 밖에 없다.

```python
# 0 ~ 1 사이의 값 100개를 표준편차를 기반으로 값을 조정한다.
x = np.random.normal(0, 1, 100)
y = np.random.normal(0, 1, 100)
x2 = np.random.normal(0, 1, 100)
y2 = np.random.normal(0, 1, 100)

plt.title('산점도')
plt.scatter(x, y)
plt.scatter(x2, y2)

plt.show()
```
