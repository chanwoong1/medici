```python
from sklearn.preprocessing import OneHotEncoder, LabelEncoder
from sklearn.pipeline import make_pipeline
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import PolynomialFeatures
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression
import pandas as pd
```

```python
car_df = pd.read_excel('데이터/data4/hyundaiCar.xlsx')
le_type = LabelEncoder()
le_fuel = LabelEncoder()
le_trans = LabelEncoder()

car_df['종류'] = le_type.fit_transform(car_df['종류'])
print(le_type.classes_)
car_df['연료'] = le_fuel.fit_transform(car_df['연료'])
print(le_fuel.classes_)
car_df['변속기']= le_trans.fit_transform(car_df['변속기'])
print(le_trans.classes_)


x_data = car_df.iloc[:,1:]
y_data = car_df.iloc[:,0]
```

```python
car_df
```

```python
x_train, x_test, y_train, y_test = train_test_split(x_data, y_data, 
                                test_size=0.2,random_state=1)

model = make_pipeline( PolynomialFeatures(degree=10), StandardScaler(), LinearRegression()  )
model.fit(x_train, y_train)
```

```python
import numpy as np
test_data=pd.DataFrame(np.array([2015, '준중형', 12.3, 204, 27,'가솔린',0,1591,1300,'자동']).reshape(1,-1))
test_data.columns = car_df.columns[1:]
test_data['종류'] = le_type.transform(test_data['종류'])
test_data['연료'] = le_fuel.transform(test_data['연료'])
test_data['변속기'] = le_trans.transform(test_data['변속기'])

test_data
```

```python
y_predicted = model.predict(test_data)
y_predicted
```
