# 01. cost

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def cost(x, y, w) :
    c = 0
    for i in range(len(x)) :
        hx = w * x[i]
        c = c + (hx - y[i]) ** 2
    return c/len(x)
```

```python
x_data = [1, 2, 3]
y_data = [1, 2, 3]
```

```python
print(cost(x_data, y_data, -1))
print(cost(x_data, y_data, 0))
print(cost(x_data, y_data, 1))
print(cost(x_data, y_data, 2))
```

```python
plt.scatter(x_data, y_data, color = 'r')
plt.show()
```

# 02. Gradient

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def cost(x, y, w) :
    c = 0
    for i in range(len(x)) :
        hx = w * x[i]
        c = c + (hx - y[i]) ** 2
    return c/len(x)
```

```python
def gradient(x, y, w) :
    c = 0
    for i in range(len(x)) :
        hx = w * x[i]
        c = c + (hx - y[i]) * x[i]
    return c/len(x)
```

```python
def show (x, y, w) :
    hist = []
    ww = []
    for i in range(200) :   # 학습횟수
        c = cost (x, y, w)
        hist.append(c)
        print('i = ', i, 'cost = ', c, 'w = ', w)
        g = gradient(x, y, w)
        ww.append(w)
        w = w - 0.1 * g
    print('최종 w : ', w)
    plt.plot(ww)
```

```python
x_data = [1, 2, 3]
y_data = [1, 2, 3]
```

```python
show(x_data, y_data, 10)
```

# 03. gradient

```python
x = np.array([1, 2, 3, 4, 5])
y = np.array([5, 8, 11, 14, 17])
learning_rate = 0.01
```

```python
x + 2
```

```python
def cost(x, y, w) :
    hy = w * x
    cost = np.sum((hy - y) ** 2) / len(x)
    return cost
```

```python
x_data = np.array([1, 2, 3])
y_data = np.array([1, 2, 3])
cost(x_data, y_data, -1)
```

```python
w = 5
b = 5
n = len(x)
epochs = 1000
```

```python
hist = []
for i in range(epochs) : 
    hy = w * x + b
    cost = np.sum((hy - y) ** 2) / n
    hist.append(cost)
    gradientW = np.sum((w * x + b - y) * 2 * x) / n 
    gradientB = np.sum((w * x + b - y) * 2) / n
    w = w - learning_rate * gradientW
    b = b - learning_rate * gradientB
    print('cost : ', cost, 'w : ', w, 'b : ', b)
    plt.plot(hist)

print('최종 w : ', w)
print('최종 b : ', b)
```
